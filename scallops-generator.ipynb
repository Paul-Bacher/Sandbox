# Scallops Generator

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Grid size
WIDTH = 250
HEIGHT = 250
```

```python
def generate_scallops(start):
    """
    Generate a scallop pattern according to the starting (top) line.
    Each line is built after the one above:
    A pixel lights up if 1 or 2 is lit among the 3 above.
    If 0 or 3, the pixel stays off.
    Pixel On: 1
    Pixell Off: 0
    """
    
    # Grid
    grid = np.zeros((HEIGHT, WIDTH))
    grid[0, :] = start
    
    for i in range(1, HEIGHT):

        # Calculate counts of a row
        counts = np.zeros(WIDTH)
        for j in range(1, WIDTH - 1):
            counts[j] = grid[i-1, j-1] + grid[i-1, j] + grid[i-1, j+1]
        counts[0] = grid[i-1, 0] + grid[i-1, 1]
        counts[WIDTH - 1] = grid[i-1, WIDTH - 1] + grid[i-1, WIDTH - 2]
        counts = counts.reshape(1, WIDTH)

        # New line
        line = np.vectorize({0: 0, 1: 1, 2: 1, 3: 0}.get)(counts)

        # Update grid row
        grid[i, :] = line
        
    return grid
```

```python
def display_grid(grid):
    
    plt.figure(figsize=(WIDTH/WIDTH*20, HEIGHT/WIDTH*20))
    plt.imshow(grid, cmap='gray')
    plt.axis('off')
    plt.show()
```

# Start = Random

```python
start = np.random.randint(2, size=(1, WIDTH))
start
```

```python
grid = generate_scallops(start)
display_grid(grid)
```

# Start = [1, ..., 1]

```python
start = np.ones((1, WIDTH))
start
```

```python
grid = generate_scallops(start)
display_grid(grid)
```

# Start = [0, 1, ..., 1]

```python
start = np.zeros((1, WIDTH))
start[0][0] = 1
start
```

```python
grid = generate_scallops(start)
display_grid(grid)
```

# Start = [1, 0, 0, 0, 0, 0, 0, 0] x N

```python
start = np.array([1 if (x % 8 == 0) else 0 for x in range(WIDTH)])
start
```

```python
grid = generate_scallops(start)
display_grid(grid)
```
